```python
import pandas as pd
from fbprophet import Prophet
import datetime
```

```python
df = pd.read_csv('raw_files/sale_prices_state.csv')
df.head()
```

```python
df = df.rename(columns = {'Date':'ds', 'FL':'y'})
df.head()
```

```python
df.dtypes
```

```python
df['ds'] = pd.to_datetime(df['ds'], errors='coerce')
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
#did not do this step yet since not sure it's necessary to have interactive graph per our convo with Manny
from fbprophet.plot import plot_plotly
import plotly.offline as py
py.init_notebook_mode()

fig = plot_plotly(m, forecast)  # This returns a plotly Figure
py.iplot(fig)
```

```python
import pickle
pkl_path = "FL.pkl"
with open(pkl_path, "wb") as f:
    # Pickle the 'Prophet' model using the highest protocol available.
    pickle.dump(m, f)

# save the dataframe
forecast.to_pickle("FL.pkl")
print("*** Data Saved ***")
```
